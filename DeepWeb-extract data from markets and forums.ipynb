```python
for glo in glob.glob(f'{P6}/*'):
#    print(glo)
    mktsoup = BeautifulSoup(open(glo), 'html.parser')
#    print(mktsoup.prettify())
    with open(f'{glo[57:67]}.csv', 'a') as f:  
        writeFile = csv.writer(f)
        for vendors in mktsoup.find_all('div','item'):
            if vendors.find('div','item_title') is None:
                item = vendors.find('div','item_title')
            else:
                item = vendors.find('div','item_title').get_text()
            if vendors.find('div','price') is None:
                price = vendors.find('div','price')
            else:
                price = vendors.find('div','price').get_text()
            if vendors.find('div','shipping') is None:
                ship = vendors.find('div','shipping')
            else:
                ship = vendors.find('div','shipping').get_text()
            if vendors.find('div','rating_count') is None:
                rtingcount = vendors.find('div','rating_count')
            else:
                rtingcount = vendors.find('div','rating_count').get_text()
            if vendors.find('div','vendor') is None:
                supplier = vendors.find('div','vendor')
            elif vendors.find('div','vendor').a is None:
                supplier = vendors.find('div','vendor')
            else:
                supplier = vendors.find('div','vendor').a.get_text()
            if vendors.find('span','vendor_rating') is None:
                rating = vendors.find('span','vendor_rating')
            else:
                rating = vendors.find('span','vendor_rating').get_text()
            writeFile.writerow([item,price,ship,rtingcount,supplier,rating])
```
